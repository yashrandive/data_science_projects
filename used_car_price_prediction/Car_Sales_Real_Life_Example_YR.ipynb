## Creating Dummy Variables

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first = True)
```

```python
data_with_dummies.head()
```

## Rearranging Columns

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price','Mileage', 'EngineV', 'Brand_BMW', 'Brand_Mercedes-Benz',
       'Brand_Mitsubishi', 'Brand_Renault', 'Brand_Toyota',
       'Brand_Volkswagen', 'Body_hatch', 'Body_other', 'Body_sedan',
       'Body_vagon', 'Body_van', 'Engine Type_Gas', 'Engine Type_Other',
       'Engine Type_Petrol', 'Registration_yes']
```

```python
cols
```

```python
data_preprocessed = data_with_dummies[cols]
```

## Linear Regression Model

## Declare Inputs and the Targets

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'], axis = 1)
```

## Scale the Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size = 0.2, random_state = 365)
```

## Create the Regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets(y_train)', size = 10)
plt.ylabel('Predictions(y_hat)', size = 10)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
# PLOT FOR RESIDUALS
sns.distplot(y_train - y_hat)
plt.title('Residual PDFs', size =18)
```

```python
# R-SQUARED
reg.score(x_train,y_train)
```

## Finding Weights and Bias

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha=0.2)
plt.xlabel('Targets(y_test)', size = 18)
plt.ylabel('Predictions(y_hat_test)', size = 18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Predictions'])
df_pf.head()
```
